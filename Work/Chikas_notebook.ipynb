```python
#import dependencies
import pandas as pd 
from sqlalchemy import create_engine
```

```python
#store Train CSV into a dataframe
Train=pd.read_csv("Resources/Train-1542865627584.csv")
```

```python
#store Inpatient CSV into a dataframe
Train_Inpatient=pd.read_csv("Resources/Train_Inpatientdata-1542865627584.csv")
Train_Inpatient.head()
```

```python
Train_Outpatient=pd.read_csv("Resources/Train_Outpatientdata-1542865627584.csv")
Train_Outpatient.head()
```

```python
Train_Beneficiary=pd.read_csv("Resources/Train_Beneficiarydata-1542865627584.csv")
Train_Beneficiary.head()
```

Create new dataframe with select columns 

```python
Train_df = Train[['Provider', 'PotentialFraud']]
Train_df = Train.loc[(Train['PotentialFraud'] == 'Yes')]
Train_df = Train_df.rename (columns = {'Provider': 'provider', 'PotentialFraud': 'potential_fraud'})
Train_df.head()
```

```python
Train_Inpatient_df = Train_Inpatient[['BeneID', 'Provider']]
Train_Inpatient_df = Train_Inpatient_df.rename (columns = {'BeneID': 'beneid', 'Provider': 'provider'})
Train_Inpatient_df
```

```python
Train_Outpatient_df = Train_Outpatient[['BeneID', 'Provider']]
Train_Outpatient_df = Train_Outpatient_df.rename (columns = {'BeneID': 'beneid', 'Provider': 'provider'})
Train_Outpatient_df.head()
```

```python
Train_Beneficiary_df = Train_Beneficiary[['BeneID', 'State', 'DOB']]
Train_Beneficiary_df = Train_Beneficiary_df.rename (columns = {'BeneID': 'beneid', 'State': 'state', 'DOB': 'dob'})
Train_Beneficiary_df.head()
```

```python
fraud_inpatient_provider_df = pd.merge(Train_Inpatient_df, Train_df, how='inner', on='provider')
fraud_inpatient_provider_df
```

```python
fraud_outpatient_provider_df = pd.merge(Train_Outpatient_df, Train_df, how='inner', on='provider')
fraud_outpatient_provider_df 
```

```python
outpatient_fraud_df = pd.read_sql_query("""SELECT b.beneid, b.state, o.provider, o.potential_fraud
                        FROM train_beneficiary as b
                        LEFT JOIN outpatient_provider_fraud as o
                        ON b.beneid = o.beneid;""", con=engine)
outpatient_fraud_df.head()
```

```python
inpatient_fraud_df = pd.read_sql_query("""SELECT b.beneid, b.state, i.provider, i.potential_fraud
                        FROM train_beneficiary as b
                        LEFT JOIN inpatient_provider_fraud as i
                        ON b.beneid = i.beneid
                        WHERE i.potential_fraud = 'Yes';""", con=engine)
inpatient_fraud_df.head()
```

```python
outpatient_fraud_df.to_html('outpatient_provider_fraud.html')
```

```python
inpatient_fraud_df.to_html('inpatient_provider_fraud.html')
```

```python
#plot state with the highest number of fraud for outpatient services
outpatient_beneficiary_grouped = outpatient_fraud_df['state'].value_counts()
outpatient_beneficiary_grouped = outpatient_beneficiary_grouped.to_frame()
outpatient_beneficiary_grouped['Fraud Count Outpatient'] = outpatient_beneficiary_grouped['state']
outpatient_beneficiary_grouped['STATE'] = outpatient_beneficiary_grouped.index
outpatient_beneficiary_grouped = outpatient_beneficiary_grouped.drop(['state'], axis = 1)
outpatient_beneficiary_grouped = outpatient_beneficiary_grouped.head(20)
outpatient_beneficiary_grouped.plot(x ='STATE', y='Fraud Count Outpatient', kind = 'barh')
```

```python
#plot state with the highest number of fraud for inpatient services
inpatient_beneficiary_grouped = inpatient_fraud_df['state'].value_counts()
inpatient_beneficiary_grouped = inpatient_beneficiary_grouped.to_frame()
inpatient_beneficiary_grouped['Fraud Count Inpatient'] = inpatient_beneficiary_grouped['state']
inpatient_beneficiary_grouped['STATE'] = inpatient_beneficiary_grouped.index
inpatient_beneficiary_grouped = inpatient_beneficiary_grouped.drop(['state'], axis = 1)
inpatient_beneficiary_grouped = inpatient_beneficiary_grouped.head(20)
inpatient_beneficiary_grouped.plot(x ='STATE', y='Fraud Count Inpatient', kind = 'barh', color = 'g')
```
